```python
# Make cleaning function
# clean data and merge into dataframes for each team
# drop unnecessary columns
def clean_nfl_data(df):
    df = df.drop(df.index[df['Unnamed: 2'] == 'Bye Week'])
    
    df = df.drop(['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7','Unnamed: 8', 'Unnamed: 9', 'Expected Points',
                  'Expected Points.1', 'Expected Points.2'], axis = 1)
    df = df.drop([0]).reset_index(drop = True)
    
    # identify playoffs index and drop everything after it
    pl_count = df['Unnamed: 2'].str.contains('Playoffs').sum()
    if pl_count > 0:
        i = df.index[df['Unnamed: 2'] == 'Playoffs']
        j = len(df)-i[0]
        df = df[:-j]
        df = df.drop(['Unnamed: 2'], axis = 1)
    else:
        df = df.drop(['Unnamed: 2'], axis = 1)
    
    # rename columns
    df = df.rename(columns = {'Unnamed: 0': 'Week', 'Score': 'TmScore', 'Score.1': 'OppScore', 'Offense': 'off1stDn',
                              'Offense.1': 'offTotYd', 'Offense.2': 'offPassYd', 'Offense.3': 'offRushYd',
                              'Offense.4': 'offTO', 'Defense': 'def1stDn', 'Defense.1': 'defTotYdsAllowed',
                              'Defense.2': 'defPassYdsAllowed', 'Defense.3': 'defRushYdsAllowed', 'Defense.4': 'defTOGained'})
    
    df = df.reset_index(drop = True)
    
    return df
```

```python
# Make column and index function
def nfl_team_year(idf, team, year):
    odf = clean_nfl_data(idf)
    
    odf['Team'] = team
    odf['Year'] = year
    odf = odf.set_index(['Team', 'Year', 'Week'])
    
    return odf
```
