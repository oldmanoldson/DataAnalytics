```python
# Make cleaning function
# clean data and merge into dataframes for each team
# drop unnecessary columns
def clean_nfl_data(df):
    df = df.drop(['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 8', 'Unnamed: 9', 'Expected Points',
                  'Expected Points.1', 'Expected Points.2'], axis = 1)
    df = df.drop([0]).reset_index(drop = True)
    
    # fill NaN with 0
    df = df.fillna(0)
    
    # identify playoffs index and drop everything after it
    i = df.index[df['Unnamed: 2'] == 'Playoffs']
    j = len(df)-i[0]
    
    df = df[:-j]
    
    # rename columns
    df = df.rename(columns = {'Unnamed: 0': 'Week', 'Unnamed: 1': 'Day', 'Unnamed: 2': 'Date', 'Unnamed: 7': 'Record',
                          'Score': 'TmScore', 'Score.1': 'OppScore', 'Offense': 'off1stDn', 'Offense.1': 'offTotYd',
                          'Offense.2': 'offPassYd', 'Offense.3': 'offRushYd', 'Offense.4': 'offTO', 'Defense': 'def1stDn',
                          'Defense.1': 'defTotYdsAllowed', 'Defense.2': 'defPassYdsAllowed', 'Defense.3': 'defRushYdsAllowed',
                          'Defense.4': 'defTOGained'})
    return df
```
